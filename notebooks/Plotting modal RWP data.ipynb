```python
import numpy as np
import datetime
from matplotlib import pyplot as plt
from scipy.io import loadmat
from copy import deepcopy

%matplotlib inline
```

```python
data_file_name = '/data/rwp/wind_mode_05122018.mat'
matlab_rwp_data = loadmat(data_file_name)
```

```python
data_fields = matlab_rwp_data.keys()
print(data_fields)
```

```python
t_start = np.datetime64('2018-05-12')
delta_t = np.array([np.timedelta64(int(this_record*60*60), 's') for this_record in matlab_rwp_data['Time'][0]])
```

```python
time = t_start + delta_t
height = matlab_rwp_data['Height']
snr = matlab_rwp_data['SNR']
```

```python
my_fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(time, height,snr, vmin=-10, vmax=20)
plt.ylabel('Height AGL (km)')
plt.xlabel('Time (UTC)')
cb = plt.colorbar(mappable=pc)
cb.set_label('SNR (dB)')
```

```python
deltas = np.diff(time)
```

```python
print(deltas.min())
print(deltas.max())
```

```python
threshold = np.timedelta64(300, 's')
elements_with_jump = np.where(deltas > threshold)
```

```python
print(elements_with_jump)
```

```python
this_array_is_nan = np.ones(height.shape[0])*np.nan
```

```python
naned_snr = deepcopy(snr) #I do this because some time = means link
```

```python
for this_element in  elements_with_jump[0]:
    naned_snr[ :, this_element ] = this_array_is_nan
```

```python
my_fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(time, height,naned_snr, vmin=-10, vmax=20)
plt.ylabel('Height AGL (km)')
plt.xlabel('Time (UTC)')
cb = plt.colorbar(mappable=pc)
cb.set_label('SNR (dB)')
```
